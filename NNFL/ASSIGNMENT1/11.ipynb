```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt 
import math
import copy
path = '/content/drive/MyDrive/ML/data_q11.xlsx' 
df = pd.read_excel(path)
insts = df.to_numpy()
m= len(insts[:,0]) #No of instinces
n = len(insts[0,:]) #No of instinces
for i in range(0,n,1):
  insts[:,i] = (insts[:,i] - insts[:,i].mean())/insts[:,i].std()
y = np.random.randint(20 , size = (m,1))+1
insts = np.append(insts,y,axis = 1) # Assigning random levels for each instance
```

```python
def cal_mean(insts,n):
  m= len(insts[:,0])
  mean = np.ones(n)
  for i in range(n):
    mean[i] = insts[:,i].mean()
  return mean

def update(insts,mean,m,n,k):
  l = np.ones(m)
  for i in range(m):
    val = []
    for j in range(k):
      val.append(np.sum((insts[i,0:n]- mean[j])**2))
    l[i] = np.argmin(val) + 1
  return l
```

```python

K = 20 # No of levels

itr = 100
#L_curr = np.zeros(m)
L_prev = insts[:,-1]
for i in range(itr):
  
  mean_vec = []
  for j in range(K):
    insts_j = insts[insts[:,-1] == j+1]
    mean_j = cal_mean(insts_j,n)
    mean_vec.append(mean_j)
  L_curr = update(insts,mean_vec,m,n,K)
  insts[:,-1] = L_curr
```

```python
for j in range(n-1):
  l=1
  for i in range(K):
    i = insts[insts[:,-1]==i+1]
    x = i[:,j]
    y = i[:,j+1]
    l+=1
    plt.scatter(x,y)
  plt.title('f'+str(j)+' vs '+'f'+str(j+1))
  plt.show()
```
